# 1. Introdução

Os carros são considerados um dos meios de transporte mais seguros para se locomover, mas, ao mesmo tempo, são relacionados a diferentes acidentes e notícias de mortes relatadas pelos jornais. Porque presenciamos esse fenômeno? Encontramos a resposta quando paramos para analisar a conduta dos motoristas destes veículos, que pelo senso comum aqui utilizado, nem sempre seguem as Leis de Trânsito e ocasionam os mais diversos acidentes.

Partir do pressuposto do senso comum não é suficiente para nós, por isso, através de uma análise de dados exploratória a *Equipe Laranja 🍊* deseja descobrir qual é a maior causa dos acidentes causados por carros e veículos nas principais rodovias mineiras. Além de atestar se o que pensamos no nosso cotidiano é realmente o que acontece nestas rodovias, a maior parte dos acidentes de carro é causada por uma má conduta do veículo por parte dos motoristas?

Aprofundando-se ainda mais no tema, o que buscamos constatar é se a maior parte dos acidentes de carro e demais veículos são causados por motoristas dirigindo sob a influência do álcool.

Para conseguir os resultados desta análise iremos utilizar o dataset [“Acidentes Rodovias Federais Brasil Jan/07 a Dez/20”](https://www.kaggle.com/datasets/equeiroz/acidentes-rodovias-federais-brasil-jan07-a-jul19/data)  disponibilizado no [Datasets do Kaggle](https://www.kaggle.com/), que contém informações sobre os acidentes de veículos que aconteceram nas rodovias federais brasileiras de janeiro de 2007 à dezembro de 2020. De todos os dados disponibilizados neste dataset, utilizaremos aqueles referentes as rodovias em Minas Gerais (MG), em específico, suas rodovias principais: BR 381, BR 040, BR 116, BR 262 e BR 050.

O colaborador e autor principal do dataset em uso está nomeado no aplicativo como Eduardo Queiroz, tendo como seus principais contribuidores o João Vitor da Silva Souza e o Kerneler Kaggle.

Os dados contidos no banco de dados após explorados e análisados pela *Equipe Laranja 🍊* responderão as perguntas propostas, mas não necessariamente comprovarão as hipóteses inicialmente propostas. Diferentes descobertas foram feitas ao longo desta análise exploratória de dados.

## **4.1. A ingestão de álcool aparece como a maior causa de acidentes?**

Para responder a primeira da indagações propostas pela *Equipe Laranja 🍊*, isto é, se a ingestão de álcool é mesmo a maior causa de acidentes reportada nas principais rodovias mineira, como cotidianamente pensamos, uma consulta mais personalizada foi executada.

Para realizar a consulta necessária para adquirir a resposta dessa pergunta, nossa equipe comparou as causas de acidentes com a quantidade de acidentes relacionadas a elas - para análise pegamos as dez maiores causas de acidentes.

A seguir serão apresentados os códigos executados pela nossa equipe e um gráfico de barras, que visa, facilitar a vizualização do resultado obtido.

```python
# Agrupar o DataFrame pela coluna 'causa_acidente' e contar a quantidade de acidentes

top_causas_acidentes = principais_rodovias_MG['id'].groupby(principais_rodovias_MG['causa_acidente']).count().reset_index(name='quantidade_acidentes')

# Ordenar em ordem decrescente para obter as 10 maiores causas

top_causas_acidentes = top_causas_acidentes.sort_values(by='quantidade_acidentes', ascending=False).head(10)

# Exibir o resultado

top_causas_acidentes
```
