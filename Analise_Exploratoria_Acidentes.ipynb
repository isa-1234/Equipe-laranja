# 1. Introdução

Os carros são considerados um dos meios de transporte mais seguros para se locomover, mas, ao mesmo tempo, são relacionados a diferentes acidentes e notícias de mortes relatadas pelos jornais. Porque presenciamos esse fenômeno? Encontramos a resposta quando paramos para analisar a conduta dos motoristas destes veículos, que pelo senso comum aqui utilizado, nem sempre seguem as Leis de Trânsito e ocasionam os mais diversos acidentes.

Partir do pressuposto do senso comum não é suficiente para nós, por isso, através de uma análise de dados exploratória a *Equipe Laranja 🍊* deseja descobrir qual é a maior causa dos acidentes causados por carros nas rodovias brasileiras. Além de atestar se o que pensamos no nosso cotidiano é realmente o que acontece nas rodovias, a maior parte dos acidentes de carro é causada por uma má conduta do veículo por parte dos motoristas?

Aprofundando-se ainda mais no tema, o que buscamos constatar é se a maior parte dos acidentes de carro são causados por motoristas dirigindo sob a influência do álcool.

Para conseguir os resultados desta análise iremos utilizar o dataset [“Acidentes Rodovias Federais Brasil Jan/07 a Dez/20”](https://www.kaggle.com/datasets/equeiroz/acidentes-rodovias-federais-brasil-jan07-a-jul19/data)  disponibilizado no [Datasets do Kaggle](https://www.kaggle.com/), que contém informações sobre os acidentes de carro que aconteceram nas rodovias federais brasileiras de janeiro de 2007 à dezembro de 2020. Seu colaborador e autor principal está nomeado no aplicativo como Eduardo Queiroz, tendo como seus principais contribuidores o João Vitor da Silva Souza e o Kerneler Kaggle.

Os dados contidos no banco de dados serão essenciais para a realização desta análise e serão melhor descritos no tópico Metodologia.

# 2. Objetivo

 A *Equipe Laranja 🍊* tem por meio deste estudo o objetivo de realizar uma Análise Exploratória de Dados dos conjuntos de dados [Datatran2018 à Datatran2020](https://www.kaggle.com/datasets/equeiroz/acidentes-rodovias-federais-brasil-jan07-a-jul19/data) (os três mais recentes, dos conjuntos disponibilizados), com o fim de confirmar se a maior parte dos acidentes de trânsito são causados pela má conduta dos motoristas de carros, em destaque, se são causado pela direção sob a influência de álcool por parte destes. Através desta análise, as seguintes perguntas buscam ser respondidas:
 1. Dentre os acidentes de carro analisados, a má conduta dos motoristas foi o que ocasionou a maior parte dos acidentes?
 2. A ingestão de álcool aparece como a maior causa de acidentes?
 3. Qual causa de acidente de carro ocasionou mais mortes?
 4. Qual a causa de acidentes mais frequente?

**OBSERVAÇÕES:**
* `classificacao_acidente` vai trazer a classificação do acidente de acordo com a situação das vítimas: Sem Vítimas; Com Vítimas Feridas e Com Vítimas Fatais;
* `fase_dia` informa a fase do dia no momento da acidente: Amanhecer; Pleno dia; Anoitecer e Plena Noite;
* `sentido_via` traz o sentido da via em que ocorreu o acidente: Crescente ou Decresente;
* `tipo_pista`vai trazer o tipo de pista em que ocorreu o acidente, e consequentemente, um pouco sobre o sentido de circulação dos veículos, pista: Simples; Dupla e Múltipla;
* `tracado_via`informa o traçado da via em que ocorreu o acidente: Reta; Curva ou Cruzamento;
* `uso_solo` traz o tipo de solo no local do acidente: Urbano ou Rural.
